# The tqdm package

```python
# Using tqdm for tracking running time

from tqdm import tqdm
import time
for i in tqdm(range(20)):
    time.sleep(0.1)
```

```python
# Using tqdm for notebooks

from tqdm.notebook import tqdm_notebook
import time

for i in tqdm_notebook(range(10), desc = 'Progress using tqdm_notebook()'):
    time.sleep(0.1)
```

### Try/Except/Raise/Finally conditions for Errors

```python
try:
  print(y)
except NameError:                       #Note that we can define the except error
  print("Variable x is not defined")
except:                                 #Here is in case of general error, where we do not specify a specific type of error
  print("Something else went wrong")
```

```python
x = "hello"

if not type(x) is int:
  raise Exception("Only integers are allowed")     #Here we raise ourselves an Exception to be printed
  # raise TypeError("Only integers are allowed")
```

```python
try:
  print(y)
except NameError:
  print("Something went wrong")
finally:                                        # The finally kind of performing somethong any how no matter what happened beforehand
  print("The 'try except' is finished")
```

### Stop a function by return (in comparison to break in loops)

```python
def main():
    length = int(input('Length: '))
    width = int(input('Width: '))

    if length <= 0:
        print("length is not positive")
        return

    if width <= 0:
        print("width is not positive")
        return

    area = length * width
    print("The area is ",  area)

main()
```

### The random.sample function and the .index function 

```python
import random

array = [0]*6
fruits = ["Apple", "Banana", "Peach", "Orange", "Durian", "Papaya"]
j=0
while j <= 100000:
    salad = random.sample(fruits, 1)
    i = fruits.index(salad[0])
    array[i]+=1
    j+=1

array
```

```python
plt.title("$Bars~\omega$")
plt.bar(fruits, array)
plt.show()
```

## Loops
$\bold{exit}$ will stop your program no matter where you call it. <br>
$\bold{return}$ will return from a function (it will stop the specific function only) <br>
$\bold{break}$ will stop the current "while" or "for" loop <br>
$\bold{continue}$ will stop the current iteration of the current "while" or "for" loop <br>

```python
txt = 'hello world'
for cr in txt:
    if cr == ' ':
        continue
    if cr == 'r':
        break
    print(cr)

print('done')
```

### Lists

```python
# Note the difference between aplit and strip

words = "A r i e l".split(' ')
print(words)
words = "A r i e l".strip(' ')
print(words)
```

```python
words = ['apple', 'banana', 'cat']

# [].insert 
words.insert(2, 'zebra')
print(words) 

# [].append 
words.append('zebra')
print(words) 

# [].remove 
words.remove('zebra')
print(words)

# [].pop
words.pop(0)
print(words)

# [].sort() working on the same list or sorted([]) genarating a new list
words.sort()
print(words)

# .join
w = '#' + ' #'.join(words)
print(w)
```

```python
# enumerate

planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn']
for idx, planet in enumerate(planets):
    print(idx, planet)
```

```python
planets1 = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn']
planets2 = ['Mercury', 'Earth', 'Venus', 'Mars', 'Jupiter', 'Saturn']

print(sorted(planets1, key=len))
# ['Mars', 'Venus', 'Earth', 'Saturn', 'Mercury', 'Jupiter']
print(sorted(planets2, key=len))
# ['Mars', 'Earth', 'Venus', 'Saturn', 'Mercury', 'Jupiter']

print(sorted(planets1, key=lambda w: (len(w), w)))
```

#### Dictionary

```python
# Printing a dictionary and using the delete key
user = {
    'fname': 'Foo',
    'lname': 'Bar',
    'email': 'foo@bar.com',
}

for key, value in user.items():
    print("key = {}  value = {}".format(key,value))


print(user) # {'lname': 'Bar', 'email': 'foo@bar.com', 'fname': 'Foo'}

fname = user['fname']
del user['fname']
print(fname) # Foo
print(user) # {'lname': 'Bar', 'email': 'foo@bar.com'}

lname_was = user.pop('lname')
print(lname_was) # Bar
print(user) # {'email': 'foo@bar.com'}
```
